## Another data split

We then split the training data again into the actual training data and the valuation data.
This way, we can learn about the dataset from validation data without having any knowledge about the testing data.
The test way must only be used in the end to check if the model generalizes.

> This is similar to learning for an exam: the training data is the content and the validation data would be a mock exam.
If you know the training data by heart, you will do very well on the training data.
If you know the training data by heart, you probably will not be excellent in the mock exam.
You can't test your knowledge against the test data because the final exam is not known during the training phase.

```python
# Split into train and devtest
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    train_df, train_size=0.80, stratify=train_df["label"], random_state=12
)
```

The math we're using does not support strings, so we have to __encode__ the labels into a numerical representation.

We know that the scales of the input data is not normalized or standardized, so we scale them.

```python
# Use StandardScaler for the features
from sklearn.preprocessing import StandardScaler

mass_scaler = StandardScaler().fit(train_df["mass"].to_numpy().reshape(-1, 1))
height_scaler = StandardScaler().fit(train_df["height"].to_numpy().reshape(-1, 1))
```

```python
def preprocess(df):
    """
    Receives a dataframe and returns a numpy array with the training features
    """
    scaled_mass = mass_scaler.transform(df["mass"].to_numpy().reshape(-1, 1))
    scaled_height = height_scaler.transform(df["height"].to_numpy().reshape(-1, 1))
    return np.concatenate((scaled_mass, scaled_height), axis=1)
```

So, for a linear function, we need to find the coefficients a and b and the bias term, such that:

$$f(x_1, x_2) = β_1 * x_1 + β_2 * x_2 + β_3 $$

Is $ 1 $ if it its a banana, and $ 0 $ if it's an apple.

```python
# Define the linear function: two features (x1, x2) and three trainable parameters (b1, b2, b3)
def linear(x1, x2, b1, b2, b3):
    return b1 * x1 + b2 * x2 + b3
```

```python
linear(2.68, 1.19, b1=1, b2=1, b3=1)
```


This linear function $ f $ is _unbounded_, so we put it through the _sigmoid_ function such that the output is forced to be in the interval $(0, 1)$.

$$ sigmoid(x) = \frac{1}{1 + e^{-x}} $$

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
plt.plot(x, y, label="sigmoid(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
def model(x1, x2, b1, b2, b3):
    return sigmoid(linear(x1=x1, x2=x2, b1=b1, b2=b2, b3=b3))
```

```python
probabilities = model(x1=X_train[:, 0], x2=X_train[:, 1], b1=0, b2=0, b3=0)
print(probabilities[:10])
```

```python
probabilities = model(x1=X_train[:, 0], x2=X_train[:, 1], b1=1, b2=1, b3=0)
print(probabilities[:10])
```

We interpret the output $(0, 1)$ as a *probability* for being Class 1.
So, if this probability is above $0.5$, we select Class 1, otherwise Class 0.

```python
def apply_threshold(probabilities, threshold=0.5):
    return (probabilities > threshold).astype(int)
```

```python
predictions = apply_threshold(probabilities, threshold=0.5)
print(predictions[:10])
```

```python
# Compute the accuracy
def compute_accuracy(y_true, y_pred):
    return (y_pred == y_true).sum() / len(y_true)
```

```python
# Get the default accuracy:
print(compute_accuracy(y_true=y_train, y_pred=predictions))
```

```python
# Try out a different parameter set ON THE TRAIN SET and compute the accuracy.
def run_try(b1, b2, b3):
    y_proba = model(x1=X_train[:, 0], x2=X_train[:, 1], b1=b1, b2=b2, b3=b3)
    y_pred = apply_threshold(y_proba, threshold=0.5)
    return compute_accuracy(y_true=y_train, y_pred=y_pred)
```

## Decision Boundary

```python
from matplotlib.colors import Normalize, ListedColormap
import matplotlib.patches as mpatches

norm = Normalize(
    vmin=enc.transform(enc.classes_).min(), vmax=enc.transform(enc.classes_).max()
)
label_colors = ["#F7330E", "#5E96C3"]
cmap = ListedColormap(label_colors)


def draw_decision_boundary(features, labels, b1, b2, b3):
    x_min = features[:, 0].min() - 1
    x_max = features[:, 0].max() + 1
    y_min = features[:, 1].min() - 1
    y_max = features[:, 1].max() + 1

    # This generates all combinations for all (x, y) coordinates
    # The 0.01 defines the resolution: This will take a while!
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, 0.02),
        np.arange(y_min, y_max, 0.02),
    )

    Z = apply_threshold(model(x1=xx, x2=yy, b1=b1, b2=b2, b3=b3))

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)

    # Plot meshes
    fig, ax = plt.subplots()

    ax.contourf(
        xx,
        yy,
        Z,
        alpha=0.5,
        cmap=cmap,
    )

    # Plot also the training points
    ax.scatter(
        features[:, 0],
        features[:, 1],
        c=norm(labels),
        s=50,
        edgecolor="k",
        cmap=cmap,
    )

    ax.set_title(f"Linear Decision Boundary")
    ax.set_xlabel("Mass")
    ax.set_ylabel("Height")
    handles = [
        mpatches.Patch(color=cmap(norm(label_id)), label=label_name)
        for label_id, label_name in zip(enc.transform(enc.classes_), enc.classes_)
    ]
    ax.legend(handles=handles, title="Fruits")

    plt.show()
```

# But how can we find the parameters automatically?

First, we need to define a **loss** (or **error** or **cost**) function. This functions defines _how far the model is from the ground truth__.

E.g. the _absolute error_:
$$
    AE = | y - \hat{y} | = | y - model(X) |
$$

(We do this for each training sample and then compute the average error.)

```python
def mean_absolute_error(y_true, y_pred):
    return np.abs(y_true - y_pred).mean()
```

```python
mean_absolute_error(
    y_true=y_train, y_pred=model(X_train[:, 0], X_train[:, 0], b1=1, b2=1, b3=0)
)
```

This means, that on average, the model is $0.73$ away from the correct solution.

Also, there is the _Squared Error_, which has an higher value, if the model is further away from the target.

$$
    SE = (y - \hat{y})^2
$$

```python
def mean_squared_error(y_true, y_pred):
    return np.square(y_true - y_pred).mean()
```

```python
mean_squared_error(
    y_true=y_train, y_pred=model(X_train[:, 0], X_train[:, 0], b1=1, b2=1, b3=0)
)
```

We can see, that the smallest value it can reach is $0$ without an upper limit. So, we want to find the _minimum_ of this loss function.

How can we get to the minimum?

We use the _derivative_ of the loss function::

$$
    loss(b_1, b_2, b_3) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \text{model}(X))^2
$$

We fill in the model as $model(X) = b_1 * x_1 + b_2 * x_2 + b_3$ (and ignore the sigmoid for the moment):

$$
loss(b_1, b_2, b_3)
    = \frac{1}{N} \sum_{i=1}^{N} (y_i - (b_1 x_1 + b_2 x_2 + b_3))^2
$$

And with a bit of math, we get to the partial derivatives for each $b$ to see it's influence into the output.

We ignore the _sigmoid_ part above, and use __only the linear part__. Derivatives are hard!

```python
def model(x1, x2, b1, b2, b3):
    # Only the linear model without the sigmoid
    return linear(x1=x1, x2=x2, b1=b1, b2=b2, b3=b3)
```

```python
def partial_derivative_b1(x1=None, x2=None, y_true=None, b1=None, b2=None, b3=None):
    return -2 * (x1 * (y_true - (b1 * x1 + b2 * x2 + b3))).sum() / len(x1)


def partial_derivative_b2(x1=None, x2=None, y_true=None, b1=None, b2=None, b3=None):
    return -2 * (x2 * (y_true - (b1 * x1 + b2 * x2 + b3))).sum() / len(x1)


def partial_derivative_b3(x1=None, x2=None, y_true=None, b1=None, b2=None, b3=None):
    return -2 * (y_true - (b1 * x1 + b2 * x2 + b3)).sum() / len(x1)
```

```python
# What are partial derivative for X_train and an initial guess of b1=1, b2=1, b3=0?
b1, b2, b3 = 1, 1, 0

pd_b1 = partial_derivative_b1(
    x1=X_train[:, 0],
    x2=X_train[:, 1],
    y_true=y_train,
    b1=b1,
    b2=b2,
    b3=b3,
)

pd_b2 = partial_derivative_b2(
    x1=X_train[:, 0],
    x2=X_train[:, 1],
    y_true=y_train,
    b1=b1,
    b2=b2,
    b3=b3,
)

pd_b3 = partial_derivative_b3(
    x1=X_train[:, 0],
    x2=X_train[:, 1],
    y_true=y_train,
    b1=b1,
    b2=b2,
    b3=b3,
)

print(f"""
For {b1 = :.2f}: {pd_b1 = :>5.2f}
For {b2 = :.2f}: {pd_b2 = :>5.2f}
For {b3 = :.2f}: {pd_b3 = :>5.2f}
""")
```

This means, that the (average) **slope** for `b1` is positive with a value of $1.65$, slightly positive for `b2` with a value of $0.05$, and negative for `b3` with a value of $-1.00$.

We want to find a **Minimum**, so we follow the **slope** **downwards** by decreaseing the invidiual parameter by a small factor of it's partial derivative.

This small factor is called the **learning rate** ($\alpha$).

```python
alpha = 0.1
```

```python
b1, b2, b3 = 1, 1, 0
```

```python
initial_mse = mean_squared_error(
    y_true=y_train,
    y_pred=model(x1=X_train[:, 0], x2=X_train[:, 1], b1=b1, b2=b2, b3=b3),
)
print(f"The initial MSE is: {initial_mse:.4f}")
```

```python
new_b1 = b1 - alpha * pd_b1
new_b2 = b2 - alpha * pd_b3
new_b3 = b3 - alpha * pd_b3
```

```python
next_mse = mean_squared_error(
    y_true=y_train,
    y_pred=model(x1=X_train[:, 0], x2=X_train[:, 1], b1=new_b1, b2=new_b2, b3=new_b3),
)
print(f"The next MSE is: {next_mse:.4f}")
```

```python
new_b1, new_b2, new_b3
```

We can see an improvment!

Typically, one such iteration is called an **epoch** and we can automate this.
If we go downwards the partial derivatives, this is also called __Gradient Descent__.

```python
def gradient_descent(num_epochs, alpha):
    # We start with random values for b1, b2, b3
    # The randomness is fixed, so that the code behaves equally everywhere.
    b1, b2, b3 = np.random.default_rng(seed=23).uniform(low=-5, high=5, size=(3,))

    # We get the predictions from the model with this initial guess
    y_pred = model(x1=X_train[:, 0], x2=X_train[:, 1], b1=b1, b2=b2, b3=b3)

    # We compute the loss and accuracy for this initial guess
    # Note, that don't optimize the accuracy directly: We only care about the loss!
    train_losses = [mean_squared_error(y_true=y_train, y_pred=y_pred)]
    train_accuracies = [
        compute_accuracy(y_true=y_train, y_pred=apply_threshold(y_pred))
    ]

    # To check for overfitting, also check the model's performance on the validation set
    # Note, we don't to this on the test set!
    y_pred = model(x1=X_val[:, 0], x2=X_val[:, 1], b1=b1, b2=b2, b3=b3)
    val_losses = [mean_squared_error(y_true=y_val, y_pred=y_pred)]
    val_accuracies = [compute_accuracy(y_true=y_val, y_pred=apply_threshold(y_pred))]

    # Run the optimization for the given number of **epochs**.
    for epoch in tqdm(range(1, num_epochs + 1), desc="Training"):
        # Compute partial derivatives
        pd_b1 = partial_derivative_b1(
            x1=X_train[:, 0], x2=X_train[:, 1], y_true=y_train, b1=b1, b2=b2, b3=b3
        )
        pd_b2 = partial_derivative_b2(
            x1=X_train[:, 0], x2=X_train[:, 1], y_true=y_train, b1=b1, b2=b2, b3=b3
        )
        pd_b3 = partial_derivative_b3(
            x1=X_train[:, 0], x2=X_train[:, 1], y_true=y_train, b1=b1, b2=b2, b3=b3
        )

        # Update parameters (weights)
        b1 -= alpha * pd_b1
        b2 -= alpha * pd_b2
        b3 -= alpha * pd_b3

        # Predict with the newly found values and record the loss/accuracy
        # for the train set and the validation set.
        y_pred = model(x1=X_train[:, 0], x2=X_train[:, 1], b1=b1, b2=b2, b3=b3)
        mse = mean_squared_error(y_true=y_train, y_pred=y_pred)
        accuracy = compute_accuracy(y_true=y_train, y_pred=apply_threshold(y_pred))
        train_losses.append(mse)
        train_accuracies.append(accuracy)

        y_pred = model(x1=X_val[:, 0], x2=X_val[:, 1], b1=b1, b2=b2, b3=b3)
        mse = mean_squared_error(y_true=y_val, y_pred=y_pred)
        accuracy = compute_accuracy(y_true=y_val, y_pred=apply_threshold(y_pred))
        val_losses.append(mse)
        val_accuracies.append(accuracy)

        # End of optimization loop

    return (train_losses, val_losses), (train_accuracies, val_accuracies), (b1, b2, b3)
```

```python
losses, accuracies, params = gradient_descent(num_epochs=5000, alpha=0.001)
```

```python
print(f"The best parameters: {params}")
```

Typically, this is a binary classification, so we would have used `LogisticRegression` here:

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_val)
print(y_pred)
```

```python
print(classification_report(y_true=y_val, y_pred=y_pred))
```

The difference is here, that the coefficients are now ordered as a matrix $m$:
$$
[b_1, b_2]
$$
and the linear part is not
$$ \hat{y} = b_1 x_1 + b_2 x_2 + b_3 $$
but rather a matrix multiplication:
$$ \hat{y} = m x + b $$ (and the sigmoid function around it)

```python
y_pred = apply_threshold(sigmoid(X_val @ clf.coef_.T + clf.intercept_).reshape((-1,)))
```

```python
print(classification_report(y_true=y_val, y_pred=y_pred))
```
